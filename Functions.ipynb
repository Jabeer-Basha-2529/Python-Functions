# Theory Questions
1. **What is the difference between a function and a method in python?**
- Function: A block of reusable code that is defined using the def keyword and is not bound to any object. It can be called independently.
  - Example: len(), sum(),  print()
- Method: A function that is associated with an object (belongs to a class). It is called on an object using dot notation.
  - Example: list.append(), string.upper(), dict.keys()
- Key Difference: Methods are functions defined inside a class and receive self as the first parameter automatically.

2. ** Explain the concept of function arguments and parameters in Python.**
- Parameters: Variables listed inside the function definition(in parentheses). They act as placeholders.
  - Example: def add(a, b): --> a and b are parameters.
- Arguments: Actual values passed to the function when it is called.
  - Example: add(5, 10) --> 5 and 10 are arguments.
Typed of Arguments:
- Positional arguments
- Keyword arguments (add(a=5, b=10) )
- Default arguments (def add(a, b=0) )
- Variable-length arguments ( *args, **kwargs)

3. What are the different ways to define and call a function in Python?
- Ways to define a function:
  1. Regular function: def myfunc():
  2. Lamba function: lamba x: x*2
  3. Nested function
  4. Function inside class(Method)
- Ways to call:
  1. Positional: func(10, 20)

  4. As a callback or passed to another function(map, filter, etc)

- If no return is used, the function returns None by default.
- You can return multiple values using tuples.
 Example:

```
     return a + b

```

5. What are iterators in python and how do they differ from iterables?
- Iterable: An object that can return an iterator (has __iter__() method).
- Iterator: An object that has both __iter__() and __next__()methods. It maintains state and produces items one by one.
 Difference:
 - All iterators are iterables, but not all iterables are iterators.
 - You can call iter() on an itrerable to get an iterator.
Example:
```
my_list = [1,2,3]   #Iterable
it = iter(my_list)  #Iterator
print(next(it))     #1
```

6. Explain the concept of generators in Python and how they are defined.
Generators are special iterators that generate values lazily(on-demand) instead of storing them all in memory.
 Defining a generator:
 1. Using yield keyword instead of return
 2. Using generator expression: (x**2 for x in range(10))
Example:
```
def my_generator():
    yield 1
    yield 2
    yield 3
for num in my_generator():
    print(num)
```
7. What are the advantages of using generators over regular functions?
- Memory efficient - values are generated one at a time(lazy evaluation)
- Better performance with large or infinite sequences
- Can represent infinite streams(e.g., Fibonacci)
- Faster startup time

8. What is a lambda function in Python and when is it typically used?
Lambda function is an anonymous, single-expression function defined using the lambda keyword.
Syntax:
`lamba arguments: expression`
Typical use cases:
- Short, throwaway functions
- With map(), filter(), sorted(), reduce()
- As callbacks
Example:
```
square = lambda x: x**2
print(square(5))
```

9. Explain the purpose and usage of the map() function in Python.
 map() applies a given function to each item of an iterable and returns a map object(iterator).
 Syntax: map(function, iterable)
 Example:
 numbers = [1, 2, 3, 4]
 squares = list(map(lambda x: x**2, numbers))
 print(squares)

````python
# 1. Write a python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def even_sum(lst):
   return sum(num for num in lst if num % 2 == 0 )
print(even_sum([1,2,3,4,5,6]))
````

````python
# 2. Create a python function that accepts a string and returns the reverse of that string.
def rev_str(s):
    return s[::-1]
print(rev_str("Hello"))
````

````python
# 3. return list of squares
def square_list(lst):
    return [x**2 for x in lst]
print(square_list([1,2,3,4]))
````

````python
# 4.Check if a number is prime(1 to 100)
def is_prime(n):
    if n <= 1:
      return False
    for i in range(2, int(n**0.5) + 1):
      if n % i == 0:
         return False
    return True
primes = [n for n in range(1, 201) if is_prime(n)]
print(primes[:10])
````

````python
# 5.Iterator class for Fibonacci sequence
class FibonacciIterator:
  def __init__(self, max_terms):
      self.max_terms = max_terms
      self.a, self.b = 0 , 1
      self.count = 0
  def __iter__(self):
      return self

  def __next__(self):
      if self.count >= self.max_terms:
        raise StopIteration
      result = self.a
      self.a, self.b = self.b, self.a + self.b
      self.count += 1
      return result
````

````python
# 6.Generator for powers of 2
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for num in powers_of_2(5):
  print(num)
````

````python
# 7. Generator to read file line by line
def read_file_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()
````

````python
# 8.Lambda to sort list of tuples by second element
tuples = [(1,5), (3,2), (2,8), (4,1)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])
print(sorted_tuples)
````

````python
# 9. map() to convert Celsius to Fahrenheit
celsius = [0, 20, 37, 100]
ft = list(map(lambda c: (c* 9/5) + 32, celsius))
print(ft)
````

````python
# 10.filter() to remove vowels
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, s))
print(remove_vowels("Hello World"))
````
